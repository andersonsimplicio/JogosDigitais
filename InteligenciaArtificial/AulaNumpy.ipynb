```python
import numpy as np
import matplotlib.pyplot as plt
```

### A variância é calculada como a média dos quadrados das diferenças entre cada valor e a média do conjunto. Como estamos elevando ao quadrado essas diferenças, a variância sempre será um valor positivo, e quanto   maior a variância, mais dispersos estão os dados.

```python
duracao_marca_a = [980, 1005, 995, 1000, 1020, 990, 1000, 1010, 1000, 1005]
duracao_marca_b = [800, 1200, 950, 1050, 1000, 850, 1100, 1150, 950, 1020]

# Cálculo da média e variância para cada marca
media_a = np.mean(duracao_marca_a)
variancia_a = np.var(duracao_marca_a)

media_b = np.mean(duracao_marca_b)
variancia_b = np.var(duracao_marca_b)

# Exibindo os resultados
print(f"Marca A - Média: {media_a}, Variância: {variancia_a}")
print(f"Marca B - Média: {media_b}, Variância: {variancia_b}")

# Gráfico de Dispersão
plt.figure(figsize=(10, 5))

# Plotando a Marca A
plt.scatter(range(1, len(duracao_marca_a) + 1), duracao_marca_a, color='blue', label=f'Marca A (Var = {variancia_a:.2f})')

# Plotando a Marca B
plt.scatter(range(1, len(duracao_marca_b) + 1), duracao_marca_b, color='red', label=f'Marca B (Var = {variancia_b:.2f})')

# Adicionando a linha da média
plt.axhline(y=media_a, color='blue', linestyle='--', label='Média Marca A')
plt.axhline(y=media_b, color='red', linestyle='--', label='Média Marca B')

# Títulos e legendas
plt.title('Comparação da Dispersão das Durações de Lâmpadas (Marca A vs Marca B)')
plt.xlabel('Lâmpada')
plt.ylabel('Duração (horas)')
plt.legend()

# Exibindo o gráfico
plt.show()
```

#  Desvio padrão
<p>quantifica essa dispersão: quanto maior o desvio padrão, mais os dados se afastam da média, e menos consistente é a característica sendo medida (neste caso, a duração das lâmpadas). Por outro lado, um desvio padrão baixo indica que os dados estão mais próximos da média, sugerindo uma maior consistência.</p>

```python
duracao_marca_a = [980, 1005, 995, 1000, 1020, 990, 1000, 1010, 1000, 1005]
duracao_marca_b = [800, 1200, 950, 1050, 1000, 850, 1100, 1150, 950, 1020]
media_a = np.mean(duracao_marca_a)
desvio_padrao_a = np.std(duracao_marca_a)

media_b = np.mean(duracao_marca_b)
desvio_padrao_b = np.std(duracao_marca_b)
print(f"Marca A - Média: {media_a}, Desvio Padrão: {desvio_padrao_a}")
print(f"Marca B - Média: {media_b}, Desvio Padrão: {desvio_padrao_b}")
plt.figure(figsize=(10, 5))
plt.scatter(range(1, len(duracao_marca_a) + 1), duracao_marca_a, color='blue', label=f'Marca A (σ = {desvio_padrao_a:.2f})')
plt.scatter(range(1, len(duracao_marca_b) + 1), duracao_marca_b, color='red', label=f'Marca B (σ = {desvio_padrao_b:.2f})')
plt.axhline(y=media_a, color='blue', linestyle='--', label='Média Marca A')
plt.axhline(y=media_b, color='red', linestyle='--', label='Média Marca B')
plt.title('Comparação da Dispersão das Durações de Lâmpadas (Marca A vs Marca B)')
plt.xlabel('Lâmpada')
plt.ylabel('Duração (horas)')
plt.legend()
plt.show()
```
